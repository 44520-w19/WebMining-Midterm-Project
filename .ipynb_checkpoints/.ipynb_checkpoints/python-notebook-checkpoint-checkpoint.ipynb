```python
#from github_token import ACCESS_TOKEN
from github import Github
import networkx as nx
import matplotlib.pyplot as plt
import pickle


access_token = '22228a29e7cb96ef8006f9b53f1795954017979c'
client = Github(access_token, per_page=100)

#gets repo info for 'TheAlgorithms/Java' repo
javaRepo = client.get_repo('TheAlgorithms/Java')
#gets user login info for contributors to 'TheAlgorithms/Java'
javaContributors = javaRepo.get_contributors()
#gets repo info for 'TheAlgorithms/C'
cRepo = client.get_repo('TheAlgorithms/C')
#gets user login info for contributors to 'TheAlgorithms/C'
cContributors = cRepo.get_contributors()

#saves a list of C repo info to a file
with open('cRepo.pkl', 'wb') as f:
    pickle.dump(cRepo, f)
with open('cRepo.pkl', 'rb') as f:
    cRepo_pkl = pickle.load(f)

#saves a list of Java repo info to a file
with open('javaRepo.pkl', 'wb') as f:
    pickle.dump(javaRepo, f)
with open('javaRepo.pkl', 'rb') as f:
    javaRepo_pkl = pickle.load(f)

#saves a list of all C contributors to a file
with open('cContributors.pkl', 'wb') as f:
    pickle.dump(cContributors, f)

with open('cContributors.pkl', 'rb') as f:
    cContributors_pkl = pickle.load(f)

#saves a list of all Java contributors to a file
with open('javaContributors.pkl', 'wb') as f:
    pickle.dump(javaContributors, f)

with open('javaContributors.pkl', 'rb') as f:
    javaContributors_pkl = pickle.load(f)


#graphs 20 contributors according to repo and user
#returns user login of 20 contributors to the repo 'TheAlgorithms/Java' in list
def graph_contributors(graph, repo):
    contributors = repo.get_contributors()[:20]
    for sg in contributors:
        graph.add_node(sg.login, type='user')
        graph.add_edge(sg.login, repo.name, type='contributes')
    return contributors

#gets the repos for each user from contributors and graphs it according to language
#returns a list of repos for the user
def graph_userLang(graph, user):
    repos = user.get_repos()
    for repo in repos:
        graph.add_node(repo.name, type='repo')
        graph.add_edge(user.login, repo.name, type='owns')
        if repo.language != None:
            graph.add_node(repo.language, type='lang')
            graph.add_edge(user.login, repo.language, type='uses')
            graph.add_edge(repo.name, repo.language, type='speaks')
    return repos

#function to make graph look pretty
def graph_colors(graph):
    colormap = {'user': 'green', 'lang':'violet', 'repo': 'blue'}
    colorlist = []
    for node in graph.nodes():
        nodetype = graph.nodes.data()[node]['type']
        colorlist.append(colormap.get(nodetype, 'black'))
    return colorlist

#function to get the correct labels for most used language
def pie_chart_labels(degrees):
    pie_labels = []
    #sorts the tuple of languages and language count by second value language count
    list_degr = sorted(degrees, key=lambda tup: tup[1])
    #gets the last 5 values out of list that have the greatest value
    newList_degr = list_degr[-5:]
    #only want 5 most used languages for pie chart
    for degree in newList_degr:
        if (len(pie_labels) < 5):
            pie_labels.append(degree[0])
    return pie_labels

#function to find the correct sizes for each language in pie chart
def pie_chart_sizes(pie_labels, degrees):
    pie_sizes = []
    for labels in pie_labels:
        pie_sizes.append(degrees(labels))
    return pie_sizes

#function to calculate the diversity of each language graph
def diversity_index(indegrees):
    sumIndividual = 0
    sumTotal = 0
    bigN = 0
    diversity = 0
    for degree in indegrees:
        if (degree[1] != 0):
            #Sum of n(n-1)
            #n = number of elements for each language
            sumIndividual = sumIndividual + (degree[1] * (degree[1] - 1))
            #Sum of all elements N
            sumTotal += degree[1]
    #N(N-1)
    bigN = sumTotal * (sumTotal - 1)
    #D = 1 - (n(n-1))/(N(N-1))
    diversity = 1 - (sumIndividual/bigN)
    return diversity

#sets up pie chart for java contributors
fig1, ax1 = plt.subplots()
#sets up pie chart for c contributors
fig2, ax2 = plt.subplots()

#creates initial graph for Java repo
gJava = nx.DiGraph()
#creates initial graph for C repo
gC = nx.DiGraph()

#calls graph_contributors function returns list of 20 contributors
cContributors = graph_contributors(gC, cRepo)

#saves the data of cContributors in a file
with open('cContributors20.pkl', 'wb') as f:
    pickle.dump(cContributors, f)
with open('cContributors20.pkl', 'rb') as f:
    cContributors20_pkl = pickle.load(f)

#calls graph_contributors function returns list of 20 contributors
javaContributors = graph_contributors(gJava, javaRepo)

#saves the data of javaContributors in a file
with open('javaContributors20.pkl', 'wb') as f:
    pickle.dump(javaContributors, f)
with open('javaContributors20.pkl', 'rb') as f:
    javaContributors20_pkl = pickle.load(f)

#calls the graph_userLang function and graphs languages of Java contributors
for user in javaContributors:
    graph_userLang(gJava, user)
    
#calls the graph_userLang function and graphs languages of C contributors
for user in cContributors:
    graph_userLang(gC, user)

#finds only the nodes for user and language
java_nodes_of_interest = [node for node in gJava.nodes() if
                     gJava.nodes.data()[node]['type'] in ['user', 'lang']]

c_nodes_of_interest = [node for node in gC.nodes() if
                     gC.nodes.data()[node]['type'] in ['user', 'lang']]

#creates a sub graph of graph gJava using the nodes of interest and nodes to keep
javalang_graph = gJava.subgraph(java_nodes_of_interest)
#creates a sub graph of graph gC using the nodes of interest and nodes to keep
clang_graph = gC.subgraph(c_nodes_of_interest)
#finds how many languages have been used at each node.. degrees
javaIndegrees = javalang_graph.in_degree()
cIndegrees = clang_graph.in_degree()


#calls functions for pie chart and puts correct data into labels and sizes
java_pie_labels = pie_chart_labels(javaIndegrees)
java_pie_sizes = pie_chart_sizes(java_pie_labels, javaIndegrees)
#Pie chart for C
c_pie_labels = pie_chart_labels(cIndegrees)
c_pie_sizes = pie_chart_sizes(c_pie_labels, cIndegrees)

print('Question: Who uses more diverse programming languages?')
print('People who use Java as their main language or people who use C as their main language?\n\n')

#generates pie chart for java contributors
ax1.pie(java_pie_sizes, labels=java_pie_labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
#plt.show()

#prints labels and sizes for java pie chart to screen
print('Java pie chart details:')
print(java_pie_labels)
print(java_pie_sizes)

#generates directed graph outline
fig=plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)

#controls the size of each node based on how many times a language was used
jSizes = [(javaIndegrees[node]+1)*100 for node in javalang_graph.nodes()]
cSizes = [(cIndegrees[node]+1)*100 for node in clang_graph.nodes()]

#matches languages to repos for labeling
jLabels = {n:n for n in javalang_graph.nodes()}
cLabels = {n:n for n in clang_graph.nodes()}


#prints labels and sizes for C pie chart to screen
print('\n\nC pie chart details:')
print(c_pie_labels)
print(c_pie_sizes)


#draws lang_graph for Java contributors languages
nx.draw(javalang_graph, nx.spring_layout(javalang_graph), arrows=True, ax=ax,
       node_size=jSizes, edge_color='#aaaaaa',
       node_color=graph_colors(javalang_graph), alpha=0.8,
       font_size=12, labels = jLabels, node_shape='*')
#generates pie chart for C contributors
ax2.pie(c_pie_sizes, labels=c_pie_labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax2.axis('equal')  
plt.show()

#prints graph info for java lang_graph
print('Name: Java contributors Repos to language graph')
print(nx.info(javalang_graph))



fig3 = plt.figure(figsize=(15,15))
ax3 = fig3.add_subplot(111)

#prints graph info for c lang_graph
print('\n\nName: C contributors Repos to language graph')
print(nx.info(clang_graph))

#draws lang_graph for C contributors languages
nx.draw(clang_graph, nx.spring_layout(clang_graph), arrows=True, ax=ax3,
       node_size=cSizes, edge_color='#aaaaaa',
       node_color=graph_colors(clang_graph), alpha=0.8,
       font_size=12, labels = cLabels, node_shape='*') 

print('\n\nConclusion:\n It seems that people who code in C often tend to use slightly more diverse programming languages.')
print('Most people who code in C know how to code in Java as well, But people who primarily code in Java tend to not have as ')
print('much experience in C. This may be because people learn how to code in Java before learning C.')
print('The diversity index for users who contributed to TheAlgorithms/Java repo was : ')
print(diversity_index(javaIndegrees))
print('The diversity index for users who contributed to TheAlgorithms/C repo was : ')
print(diversity_index(cIndegrees))
print('All in all, both groups of users have high diversity scores for what types of languages they like to program in.')
```
