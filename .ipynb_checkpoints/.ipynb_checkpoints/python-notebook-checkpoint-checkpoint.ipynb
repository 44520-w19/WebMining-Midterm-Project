```python
#from github_token import ACCESS_TOKEN
from github import Github
import networkx as nx
import matplotlib.pyplot as plt
import pickle


access_token = '22228a29e7cb96ef8006f9b53f1795954017979c'
client = Github(access_token, per_page=100)
repo = client.get_repo('doocs/advanced-java')
gazers = repo.get_stargazers()
with open('javaRepo.pkl', 'wb') as f:
    pickle.dump(repo, f)
with open('javaRepo.pkl', 'rb') as f:
    javaRepo_pkl = pickle.load(f)

with open('stargazers.pkl', 'wb') as f:
    pickle.dump(gazers, f)

with open('stargazers.pkl', 'rb') as f:
    gazers_pkl = pickle.load(f)

#labels = {n:n for n in g.nodes()}
#nx.draw(g, nx.spring_layout(g), arrows=True, ax=ax,
#       node_size=100, edge_color='#aaaaaa',
#       node_color=graph_colors(g), alpha=0.8,
#       font_size=12, labels = labels, node_shape='*')


"""
language_count = 0
for i in range(20):
    #only gets 20 user login names
    user = client.get_user(gazers[i].login)
    #list of all repos for each user
    newrepos = user.get_repos()
    #the first repo in each users repos
    print(newrepos[0])
    #name of first repo
    print(newrepos[0].name)
    #users login name
    print(user.login)
    language_count = 0
    #checks all repos in each of 20 stargazers and prints their languages of each repo
    for repo in newrepos:
        if (repo.language != None):
            print(repo.language)
            #keeps count of how many repos each stargazer has
            language_count = language_count + 1
    print(language_count)
#print(dir(repos))"""
#returns user login of 20 stargazers in list
def graph_stargazers(graph, repo):
    gazers = repo.get_stargazers()[:20]
    for sg in gazers:
        graph.add_node(sg.login, type='user')
        graph.add_edge(sg.login, repo.name, type='gazes')
    return gazers

#returns a list of repos for the user
def graph_userLang(graph, user):
    repos = user.get_repos()
    for repo in repos:
        graph.add_node(repo.name, type='repo')
        graph.add_edge(user.login, repo.name, type='owns')
        if repo.language != None:
            graph.add_node(repo.language, type='lang')
            graph.add_edge(user.login, repo.language, type='uses')
            graph.add_edge(repo.name, repo.language, type='speaks')
    return repos

def graph_colors(graph):
    colormap = {'user': 'green', 'lang':'violet', 'repo': 'blue'}
    colorlist = []
    for node in graph.nodes():
        nodetype = graph.nodes.data()[node]['type']
        colorlist.append(colormap.get(nodetype, 'black'))
    return colorlist

def pie_chart_labels(degrees):
    pie_labels = []
    for degree in degrees:
        if (degree[1]>5) and (len(pie_labels) < 5):
            pie_labels.append(degree[0])
    return pie_labels

def pie_chart_sizes(pie_labels, degrees):
    pie_sizes = []
    for labels in pie_labels:
        pie_sizes.append(degrees(labels))
    return pie_sizes
fig=plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
fig1, ax1 = plt.subplots()

g = nx.DiGraph()
g.add_node(repo.name, type='repo')

javaGazers = graph_stargazers(g, repo)
with open('javagazers20.pkl', 'wb') as f:
    pickle.dump(javaGazers, f)

with open('javagazers20.pkl', 'rb') as f:
    javaGazers_pkl = pickle.load(f)

for user in javaGazers:
    graph_userLang(g, user)
    
print(nx.info(g))
    
nx.write_gpickle(g, 'bigJavaGraph')

#java_graph = nx.read_gpickle('bigJavaGraph')

nodes_of_interest = [node for node in g.nodes() if
                     g.nodes.data()[node]['type'] in ['user', 'lang']]


lang_graph = g.subgraph(nodes_of_interest)
indegrees = lang_graph.in_degree()

java_pie_labels = pie_chart_labels(indegrees)

java_pie_sizes = pie_chart_sizes(java_pie_labels, indegrees)

ax1.pie(java_pie_sizes, labels=java_pie_labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.show()

print(java_pie_labels)
print(java_pie_sizes)

"""
for degree in indegrees:
    if (degree[1]> 5):
        pie_labels.append(degree[0])
        print(pie_labels)

for labels in pie_labels:
    print(indegrees(labels))
"""

sizes = [(indegrees[node]+1)*100 for node in lang_graph.nodes()]
labels = {n:n for n in lang_graph.nodes()}
print(nx.info(lang_graph))
nx.draw(lang_graph, nx.spring_layout(lang_graph), arrows=True, ax=ax,
       node_size=sizes, edge_color='#aaaaaa',
       node_color=graph_colors(lang_graph), alpha=0.8,
       font_size=12, labels = labels, node_shape='*')

```
